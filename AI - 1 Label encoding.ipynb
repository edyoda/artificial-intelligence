```python
from sklearn import preprocessing
```

```python
# Sample input labels 
input_labels = ['red', 'black', 'red', 'green', 'black', 'yellow', 'white'] 
```

```python
encoder = preprocessing.LabelEncoder() 
encoder.fit(input_labels) 
```

```python
for i, item in enumerate(encoder.classes_):
    print (item, '-->', i)
```
